```python
import umap
# import matplotlib.pyplot as plt
# df.reset_index(inplace=True)
#drop retina column
df = df.drop(columns=['Retina'])
reducer = umap.UMAP()
embedding = reducer.fit_transform(df.iloc[:,:-2])

# Get unique values from the 'PXD' column
unique_pxd = df['PXD_folder'].unique()

# Create a color map for each unique value in 'PXD'
color_map = plt.cm.get_cmap('tab10', len(unique_pxd))

# Plot the scatter plot with colors based on 'PXD'
for i, pxd in enumerate(unique_pxd):
    mask = df['PXD_folder'] == pxd
    plt.scatter(embedding[mask, 0], embedding[mask, 1], c=color_map(i), label=pxd)

plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the dataset', fontsize=24)
plt.legend()
plt.show()
```
