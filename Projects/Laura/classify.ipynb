```python
df = pd.read_excel('trf17254-sup-0003-tables1.xlsx', sheet_name='Quantified_DIS_III')
gene_to_protein_map = pd.read_csv('idmapping_reviewed_true_2024_12_06.tsv', sep='\t')   
gene_to_protein_dict = dict(zip(gene_to_protein_map['From'], gene_to_protein_map['Entry']))

df.head()
```

```python
df['Entry'] = df['Gene_names'].map(gene_to_protein_dict)
df.drop(columns=['UID', 'Gene_names'], inplace=True)
df.set_index('Entry', inplace=True)
df.head()
```

```python
#one by one, column by column
prediction_dict={}
for i in range(len(df.columns)):
    subset = df.iloc[:, i].dropna()
    subset_minmax = pd.DataFrame((subset - subset.min()) / (subset.max() - subset.min())).T
    test = MLMarker(subset_minmax.iloc[0:1,:], binary = False, dev=False)
    prediction = MLMarker.adjusted_absent_shap_values_df(test, penalty_factor=1)
    prediction_dict[i] = prediction.sum(axis=1).head(3)

```

```python
prediction_dict
```

```python
#row wise average to get average protein expression
df['average'] = df.iloc[:,1:-2].mean(axis=1)
baseline = df[['Entry', 'average']].pivot_table(index='Entry', values='average', aggfunc='mean')
baseline['minmax_average'] = (baseline['average'] - baseline['average'].min()) / (baseline['average'].max() - baseline['average'].min())
baseline_test = baseline[['minmax_average']].T

test = MLMarker(baseline_test.iloc[0:1,:], binary = False, dev=False)
prediction = MLMarker.adjusted_absent_shap_values_df(test, penalty_factor=1)
prediction.sum(axis=1)
```
